```python
from IPython.core.display import HTML, Markdown
import itertools

from datetime import datetime, date, timedelta
import calendar

format = "%A, %B %d, %Y"
short_format = "%a %b %d"
# 1. set the first day of classes
first_day = "Wednesday, January 22, 2025"

# 2. set the last day of class_variables
last_day = "Monday, May 5, 2025"

# 3. set the recurring days of the weeks
days_of_week = ["Monday", "Wednesday"]

# 4. add in the date of the final exam
final_exam_day = None

# 4. list any exceptions for like holidays and the like
holidays = {
    "Monday, February 17, 2025": "Presidents' Day",
    "Monday, March 24, 2025": "Spring Break",
    "Wednesday, March 26, 2025": "Spring Break",
}
holidays = {datetime.strptime(k, format): v for k, v in holidays.items()}

# 5. add in any additional dates (most commonly legislative day where off cycle class meeting might happen)
# extra_days = {'Tuesday, October 11, 2022': 'Legislative Day (classes meet on Monday schedule)'}
extra_days = {
    "Tuesday, February 18, 2025": "Legislative Day (classes meet on Monday schedule)"
}
extra_days = {datetime.strptime(k, format): v for k, v in extra_days.items()}

# this part runs the script and prints to the terminal the dates
first_day_f = datetime.strptime(first_day, format)
last_day_f = datetime.strptime(last_day, format)

# generator for all the days between two dates
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days + 1)):
        yield start_date + timedelta(n)

class_dates_info = {}
class_dates = []
for d in daterange(first_day_f, last_day_f):
    if d in extra_days:
        class_dates.append(d)
    if calendar.day_name[d.weekday()] in days_of_week:
        class_dates.append(d)
```
