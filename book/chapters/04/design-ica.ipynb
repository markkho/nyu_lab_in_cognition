# In Class Activity - Evaluating Research Designs

Evaluating research designs.  In the lecture and reading this week we talked about various aspects of research design and measurement.  In particular, much of the discussion was about assessing things that possibly can affect the validity of a research study.

Psychological Science is one of the more "flashy" journals in the field of Psychology.  The journal website is [here](https://journals.sagepub.com/home/pss) where, if you are logged into NYU, you can access the PDFs of papers published by the journal.

In this in class activity you will find a paper your group thinks is interesting from a recent publication in the journal.  Then you will assess the following about the paper:

- Is the design an experimental or observational study?
- What is the main dependent measure (outcome) and what are the various independent measures (predictors)?
- Do the authors report anything about the reliability of their measurement?  Is there reason to be concerned about it?
- What are some factors that might impact the validity of the study?  Are there confounds? Artifacts? Repeated testing effects? And so on. Please discuss any of these types of issues that might pertain to your article and how serious they might be. After considering potential threats to the study's validity, do you still believe its conclusions?

You *don't* have to read the paper line by line to answer these questions typically.  Papers are organized into subsections like "Introduction", "Methods", and "Conclusion", which can help you jump around and access the relevant information.

You will summarize both the article itself and your answers to the questions above present them to the class.
